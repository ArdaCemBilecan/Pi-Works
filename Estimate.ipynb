# Getting our data ready for training.

```python
# My independent variable will be hour,day and month that's why we have to use StandartScaler
# Also I have just do it for a city. Because for other cities , We have to follow same proccess.
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
city1_time = dic0.keys()
X_train=[]
y_train = list(dic0.values())
for time in city1_time:
    split = time.split('-')
    split = list(map(int, split))
    X_train.append(split)
```

```python
city1_test_time = test0.keys()
X_test=[]
y_test = list(test0.values())
for time in city1_test_time:
    split = time.split('-')
    split = list(map(int, split))
    X_test.append(split)
```

# Train With Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(X_test.shape , X_train.shape)
print(len(y_train) , len(y_test))
```

```python
regression = LinearRegression()
regression.fit(X_train,y_train)
```

```python
y_pred = regression.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_log_error
loss = mean_squared_log_error(y_test,y_pred)
print(loss)
```

```python
import tensorflow as tf
from tensorflow.keras.models import *
from tensorflow.keras.layers import *
from keras.layers import LeakyReLU
```

```python
np_x_train = np.array(X_train)
np_x_test = np.array(X_test)
np_y_train = np.array(y_train)
np_y_test = np.array(y_test)
print(np_x_train.shape)
```

# Train just Dense Layers

```python
model = Sequential()
model.add(Dense(256,activation=tf.keras.layers.LeakyReLU(alpha=0.1),input_shape=(3,)))
model.add(Dense(256,activation=tf.keras.layers.LeakyReLU(alpha=0.1)))
model.add(Dense(64,activation=tf.keras.layers.LeakyReLU(alpha=0.1)))
model.add(Dense(1))
model.compile(loss='mean_squared_logarithmic_error',optimizer='adam',metrics=['MeanSquaredLogarithmicError'])
model.fit(np_x_train, np_y_train , epochs=50,validation_data=(np_x_test,np_y_test))
```

# Train with GRU

```python
train_x = np_x_train.reshape(-1,1,3)
test_x = np_x_test.reshape(-1,1,3)
```

```python
model = Sequential()
model.add(GRU(256,activation=tf.keras.layers.LeakyReLU(alpha=0.1),input_shape=(1,3)))
model.add(Dense(256,activation=tf.keras.layers.LeakyReLU(alpha=0.1)))
model.add(Dense(128,activation=tf.keras.layers.LeakyReLU(alpha=0.1)))
model.add(Dropout(0.25))
model.add(Dense(64,activation=tf.keras.layers.LeakyReLU(alpha=0.1)))
model.add(Dense(32,activation=tf.keras.layers.LeakyReLU(alpha=0.1)))
model.add(Dense(1))
model.compile(loss='mean_squared_logarithmic_error',optimizer='adam',metrics=['MeanSquaredLogarithmicError'])
model.fit(train_x, np_y_train , epochs=50,validation_data=(test_x,np_y_test))
```
